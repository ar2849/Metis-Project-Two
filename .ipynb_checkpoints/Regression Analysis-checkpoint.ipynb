```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression, LassoCV
from sklearn.preprocessing import PolynomialFeatures, OneHotEncoder
import seaborn as sns
%config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

```python
# load film data set
with open("Final_data.pkl", 'rb') as picklefile: 
    film_data = pickle.load(picklefile)
```

```python
film_data = film_data.dropna()
film_data
```

```python
#Split off month and year 
film_data['months'] = pd.DatetimeIndex(film_data['release_date']).month
film_data['year'] = pd.DatetimeIndex(film_data['release_date']).year
film_data
```

```python
def ohe_distributor(categorical_variable):
    '''
    Converts categorical variables using one-hot-encoder to use the feature in regression analysis
    INPUT: pandas dataframe column 
    OUTPUT: binary encoded feature columns in a pandas dataframe
    '''
    encoder = OneHotEncoder(sparse=False, drop='first')
    ohe_cv = encoder.fit_transform(categorical_variable)
    columns = encoder.get_feature_names(['distributor'])
    distributor_ohe = pd.DataFrame(ohe_cv, columns=columns, index=categorical_variable.index)
    return distributor_ohe
```

```python
def ohe_rating(categorical_variable):
    '''
    Converts categorical variables using one-hot-encoder to use the feature in regression analysis
    INPUT: pandas dataframe column 
    OUTPUT: binary encoded feature columns in a pandas dataframe
    '''
    encoder = OneHotEncoder(sparse=False, drop='first')
    ohe_cv = encoder.fit_transform(categorical_variable)
    columns = encoder.get_feature_names(['rating'])
    rating_ohe = pd.DataFrame(ohe_cv, columns=columns, index=categorical_variable.index)
    return rating_ohe
```

```python
distributor_dummies = ohe_distributor(film_data[['distributor']])
```

```python
rating_dummies = ohe_rating(film_data[['rating']])
```

```python
one_hot_encoded_features = pd.concat([distributor_dummies, rating_dummies], axis=1)
```

```python
filtered_film_data = film_data.drop(['rating', 'distributor', 'release_date'], axis=1)
filtered_film_data
```

```python
Preprocessed_film_data = pd.concat([one_hot_encoded_features, filtered_film_data], axis=1)
Preprocessed_film_data
```

```python
Preprocessed_film_data = Preprocessed_film_data.apply(pd.to_numeric, errors='coerce')
```

K-Fold Cross Validation Process

```python
X, y = Preprocessed_film_data.drop('domestic_total_gross',axis=1), Preprocessed_film_data['domestic_total_gross']
X, X_test, y, y_test = train_test_split(X, y, test_size=.2, random_state=10) #hold out 20% of the data for final testing

#this helps with the way kf will generate indices below
X, y = np.array(X), np.array(y)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state = 1)
cv_lm_r2s = [] #collect the validation results

for train_ind, val_ind in kf.split(X,y):
    
    X_train, y_train = X[train_ind], y[train_ind]
    X_val, y_val = X[val_ind], y[val_ind] 
    
    #simple linear regression
    lm = LinearRegression()

    lm.fit(X_train, y_train)
    cv_lm_r2s.append(round(lm.score(X_val, y_val), 3))

print('Simple regression scores: ', cv_lm_r2s, '\n')

print(f'Simple mean cv r^2: {np.mean(cv_lm_r2s):.3f} +- {np.std(cv_lm_r2s):.3f}')
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state = 1)
cv_lm_poly_r2s = []

for train_ind, val_ind in kf.split(X,y):    
    #poly with degree 2
    poly = PolynomialFeatures(degree=2, interaction_only=True)

    X_train_poly = poly.fit_transform(X_train)
    X_val_poly = poly.transform(X_val)

    lm_poly = LinearRegression()
    
    lm_poly.fit(X_train_poly, y_train)
    cv_lm_poly_r2s.append(round(lm_poly.score(X_val_poly, y_val), 3))
    
print('Poly scores: ', cv_lm_poly_r2s, '\n')

print(f'Poly mean cv r^2: {np.mean(cv_lm_poly_r2s)} +- {np.std(cv_lm_poly_r2s)}')
```

Processing copy of film data frame for residual analysis

```python
Preprocessed_film_data_copy = Preprocessed_film_data.copy()
X, y = Preprocessed_film_data_copy.drop('domestic_total_gross', axis=1), Preprocessed_film_data_copy['domestic_total_gross']
film_lr = LinearRegression().fit(X, y)
Preprocessed_film_data_copy['res'] = abs(film_lr.predict(X)-y)
Preprocessed_film_data_copy['predictions']= film_lr.predict(X)
```

```python
De_outlier_data = Preprocessed_film_data_copy[Preprocessed_film_data_copy.res >= 1000000]
De_outlier_data = Preprocessed_film_data_copy[Preprocessed_film_data_copy.res <= 9999999]
De_outlier_data.sort_values(['res'])
```

```python
X, y = De_outlier_data.drop(['domestic_total_gross', 'res'],axis=1), De_outlier_data['domestic_total_gross']
X, X_test, y, y_test = train_test_split(X, y, test_size=.2, random_state=10) #hold out 20% of the data for final testing

#this helps with the way kf will generate indices below
X, y = np.array(X), np.array(y)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state = 1)
cv_lm_r2s = [] #collect the validation results

for train_ind, val_ind in kf.split(X,y):
    
    X_train, y_train = X[train_ind], y[train_ind]
    X_val, y_val = X[val_ind], y[val_ind] 
    
    #simple linear regression
    lm = LinearRegression()

    lm.fit(X_train, y_train)
    cv_lm_r2s.append(round(lm.score(X_val, y_val), 3))

print('Simple regression scores: ', cv_lm_r2s, '\n')

print(f'Simple mean cv r^2: {np.mean(cv_lm_r2s):.3f} +- {np.std(cv_lm_r2s):.3f}')
```
